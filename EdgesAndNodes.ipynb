```python
import pandas as pd
import itertools
from collections import defaultdict

# Load your uploaded file (use exact filename)
df = pd.read_csv("SARS DATASET - Sheet1.csv")

# Preview to confirm
df.head()
```

```python
# Split authors by semicolon and remove extra spaces
df['author_list'] = df['authors'].fillna('').apply(
    lambda x: [a.strip() for a in x.split(';') if a.strip()]
)

# Preview the new column
df[['authors', 'author_list']].head()
```

```python
edge_weights = defaultdict(int)

# Count how many times each pair of authors has co-authored
for authors in df['author_list']:
    for pair in itertools.combinations(sorted(set(authors)), 2):
        edge_weights[pair] += 1

# Convert to DataFrame
edges = pd.DataFrame([(a, b, w) for (a, b), w in edge_weights.items()],
                     columns=["source", "target", "weight"])

# Preview the first few rows
edges.head()
```

```python
# Extract all unique authors
unique_authors = set()
for authors in df['author_list']:
    unique_authors.update(authors)

# Create nodes DataFrame
nodes = pd.DataFrame({"id": list(unique_authors), "label": list(unique_authors)})

# Preview nodes
nodes.head()
```

```python
# Save CSV files
edges.to_csv("edges.csv", index=False)
nodes.to_csv("nodes.csv", index=False)

# Download them
from google.colab import files
files.download("edges.csv")
files.download("nodes.csv")
```
